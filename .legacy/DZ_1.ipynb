```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
from itertools import *
# Just essential tools
```

```python
df = pd.read_pickle('data/KAS_USDT_5m_300_days_data.pkl')
df_1 = pd.read_pickle('data/KAS_USDT_1m_10_days_data.pkl')

df.head(3)
# df is a 5-minute dataframe, where each row is a 5-minute candle
# Same applies to df_1, but it's recent data and shorter
```

```python
print(len(df))
df.info()
# As we can see, df is very good as it doesn't contain any null values, this is because
# df is got throug API of exchange MEXC, and it would be worst nightmare for an exchange to 
# have any undefined values in their data.

# Also keep in mind that crypto exhanges work 24/7 without holidays or breaks, so data
# doesn't have any gaps

# Unfortunately when data is fetched, all values are converted into objects, whily they clearly are
# intended to be floats 
```

```python
df = df.astype({'high': float, 'low': float, 'open': float, "close": float, 'volume': float})
df.describe()

# Now after type conversion (which could be done at the moment of import) we can get basic 
# understanding of our dataframe. 
```

```python
keys = ['high', 'low', 'open', 'close', 'volume']
fig, axes = plt.subplots(ncols=len(keys))
fig.set_size_inches(8 * len(keys), 4)
axes = axes.flatten()

for key, ax in zip(keys, axes):
    ax.hist(df[key], density=True, bins=23)
    
    loc = df[key].mean()
    scale = df[key].std()
    
    x_left, x_right = ax.get_xlim()
    x = np.linspace(x_left, x_right, 10000)
    y = stats.norm.pdf(x, loc=loc, scale=scale)
    
    ax.plot(x, y, linestyle='dashed')
    ax.set_title(key)

# As we can clearly see there's no normal distribution, and that is to be expected. It would be strange
```

```python
# IDEA: make column of % differences, and then make correlations
"""
This seemingly random idea comes not only from desparation, but also from a fact that prosentages
(in a way) are normalized by themselves
"""

df['o-c_change'] = round(((df['open'] / df['close']) - 1) * 100, 2) # 2 is arbitrary
df['h-l_change'] = round(((df['high'] / df['low']) - 1) * 100, 2) # 2 is arbitrary
median_volume = df.loc[df['volume'] != 0, 'volume'].median()
df['volume'] = df['volume'].replace(0, median_volume)
df['v_change'] = round(((df['volume'][2:-2] / df['volume'][2:-2].shift(-1)) - 1) * 100, 2) # 2 is arbitrary
df['v_change'] = df['v_change'].fillna(df['v_change'].median())
"""
Now we are getting somewhere, let's see correlation and boxplots, maybe even normalization?
"""
df.describe()
```

```python
"""
let's calculate regression value to detemine if there is dependance between open - close and 
high - low changes
"""
def sum_of_squares(samples: np.ndarray) -> float:
    return ((samples - samples.mean()) ** 2).sum()
ones = np.ones((17300,1))
x_pot = np.hstack((ones, df['o-c_change'].values.reshape(17300, 1)))
XTX = x_pot.T.dot(x_pot)
XTX_inv = np.linalg.inv(XTX)

b = XTX_inv.dot(x_pot.T).dot(df['h-l_change'])
z = x_pot.dot(b)
e = df['h-l_change'] - z

R = 1 - sum_of_squares(e) / sum_of_squares(np.array(df['h-l_change']))
print(R)

"""
Yep, no linear correlation whatsoever, values below 0.05 are tipicly considered statisticly 
insegnificant
"""
```

```python
n = 17280
m = 1
k1 = m
k2 = n - m - 1
alpha = 0.05

t = stats.f.ppf(1 - alpha, k1, k2)

F = (R / k1) / ((1 - R) / k2)
print(t, F)

"""
But when we perform the F test we can see that our model is signifficant at 5% lvl, that means that
correlation is likely enough for us to concider it. Le's try to visualize it.
"""
```

```python
b1 = np.cov(df['o-c_change'], df['h-l_change'], ddof=1)[0, 1] / np.var(df['o-c_change'], ddof=1)
b0 = df['h-l_change'].mean() - b1 * df['o-c_change'].mean()

plt.scatter(df['h-l_change'], df['o-c_change'])
x = np.linspace(0, 10, 1000)
y = b0*x + b1
plt.plot(x, y, color = 'r')
plt.show()

"""
Thst's pretty interesting, it suggests that open - close changes tilt slightly to positive values 
which means that value of a coin rises on average, which is not good indicator at all, but 
interesting suggestion nevertheless
"""
```
